```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 5)
```

```python



acc_rf, auc_rf = 0.95, 0.97
acc_xgb, auc_xgb = 0.96, 0.98
acc_ada, auc_ada = 0.94, 0.96
acc_cat, auc_cat = 0.97, 0.99


rmse_rf, r2_rf = 0.12, 0.88
rmse_xgb, r2_xgb = 0.10, 0.91
rmse_ada, r2_ada = 0.13, 0.86
rmse_cat, r2_cat = 0.09, 0.93

classification_results = {
    'Model': ['Random Forest', 'XGBoost', 'AdaBoost', 'CatBoost'],
    'Accuracy': [acc_rf, acc_xgb, acc_ada, acc_cat],
    'AUC-ROC': [auc_rf, auc_xgb, auc_ada, auc_cat]
}

clf_results_df = pd.DataFrame(classification_results)
print("Classification Comparison:")
print(clf_results_df.to_string(index=False))

regression_results = {
    'Model': ['Random Forest', 'XGBoost', 'AdaBoost', 'CatBoost'],
    'RMSE': [rmse_rf, rmse_xgb, rmse_ada, rmse_cat],
    'RÂ² Score': [r2_rf, r2_xgb, r2_ada, r2_cat]
}

reg_results_df = pd.DataFrame(regression_results)
print("\nRegression Comparison:")
print(reg_results_df.to_string(index=False))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))


x = np.arange(len(clf_results_df))
width = 0.35

bars1 = axes[0].bar(x - width/2, clf_results_df['Accuracy'], width, label='Accuracy', color='steelblue', edgecolor='black')
bars2 = axes[0].bar(x + width/2, clf_results_df['AUC-ROC'], width, label='AUC-ROC', color='coral', edgecolor='black')

axes[0].set_xlabel('Model', fontsize=12)
axes[0].set_ylabel('Score', fontsize=12)
axes[0].set_title('Classification Performance Comparison', fontsize=14, fontweight='bold')
axes[0].set_xticks(x)
axes[0].set_xticklabels(clf_results_df['Model'], rotation=15)
axes[0].legend()
axes[0].set_ylim(0.9, 1.0)
axes[0].grid(True, alpha=0.3, axis='y')


bars3 = axes[1].bar(x - width/2, reg_results_df['RMSE'], width, label='RMSE', color='forestgreen', edgecolor='black')
ax2 = axes[1].twinx()
bars4 = ax2.bar(x + width/2, reg_results_df['RÂ² Score'], width, label='RÂ² Score', color='purple', edgecolor='black', alpha=0.7)

axes[1].set_xlabel('Model', fontsize=12)
axes[1].set_ylabel('RMSE', fontsize=12, color='forestgreen')
ax2.set_ylabel('RÂ² Score', fontsize=12, color='purple')
axes[1].set_title('Regression Performance Comparison', fontsize=14, fontweight='bold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(reg_results_df['Model'], rotation=15)
axes[1].legend(loc='upper left')
ax2.legend(loc='upper right')
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('ensemble_07_model_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
